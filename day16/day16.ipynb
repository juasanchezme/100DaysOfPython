# OOP


## Procedural Programming
Concept:
Procedural programming is a programming paradigm that is based on procedures or routines (also known as functions). Its structured around the concept of procedures or functions that operate on data. The focus is on how to perform tasks by writing procedures, which are step-by-step instructions to be executed in a specific sequence.

## Object-Oriented Programming (OOP)
Concept:
Object-Oriented Programming (OOP) is a paradigm based on objects, which are instances of classes. These objects contain both data (attributes) and behavior (methods). The focus is on what to do and how objects should interact with each other, rather than focusing on the flow of the program.

```python
# Procedural approach
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def main():
    a = 5
    b = 10
    print(f"Addition: {add(a, b)}")
    print(f"Subtraction: {subtract(a, b)}")
    print(f"Multiplication: {multiply(a, b)}")

main()
```

```python
# OOP approach
class Calculator:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def add(self):
        return self.x + self.y
    
    def subtract(self):
        return self.x - self.y
    
    def multiply(self):
        return self.x * self.y

def main():
    calc = Calculator(5, 10)
    print(f"Addition: {calc.add()}")
    print(f"Subtraction: {calc.subtract()}")
    print(f"Multiplication: {calc.multiply()}")

main()
```
