## Args

```python
def add(*args):
    print(args)

add(2, 3, 8, 7)
```

```python
# por convencion se llama args pero puede ser * seguido de cualquier nombre 

def add(*numbers):
    number_list = [n for n in numbers]
    print(sum(number_list))

# el argumento * colecciona todos los elementos que pasemos como una tupla que la funcin add operar
add(1,2,3,4,5)
```

son arguments posicionales porque la posicion en la que pasamos el argumento es importante
numbers[0] = 1
numbers[1] = 2
...

```python
def calculate(** kwargs):
    print(kwargs)
    print(f"se suma {kwargs['add']}, se multiplica {kwargs['multiply']}")
    
calculate(add = 3, multiply = 5)
```

```python
def calculate(n, ** kwargs):
    n += kwargs["add"]
    n *= kwargs["multiply"]
    print(n)
    
    
calculate(n = 5, add = 3, multiply = 2)
```

we can use **.get()** for dictionaries, it's going to look for de value and it
it doesn't exist will return "none", also we can especify a specific default value if it doesn't exists

```python
class Car:
    def __init__(self, **kw):
        self.brand = kw.get("brand", "renault")
        self.color = kw.get("color", "white")
        self.year = kw.get("year")
        
my_car = Car(brand = "Toyota", color= "red")
print(my_car.brand, my_car.color, my_car.year)
        
```
