the **init** function is going to be called every time we create an object 

```python
class User:
    def __init__(self, name, id):
        self.name = name
        self.id = id
        # default attribute
        self.followers = 0

usuario = User("juan", 1)

usuario2 = User("Jose", 2)

print(usuario.name, usuario.id)
print(usuario2.name, usuario2.id)

usuario.email = "juasan@"

print(usuario.email)
```

Usiing methods

```python
class User:
    def __init__(self, name, id):
        self.name = name
        self.id = id
        # default attribute
        self.followers = 0
        self.following = 0
    
    def follow(self, user):
        user.followers += 1
        self.following += 1
        

user_1 = User("juan", "001")
user_2 = User("Jose", "002")

user_1.follow(user_2)

print(user_1.followers, user_2.following)
```
